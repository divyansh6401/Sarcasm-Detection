```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import preprocessing
from tensorflow.keras import optimizers,losses
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing import text
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM,Embedding,Flatten,SimpleRNN
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/drive/MyDrive/Sarcasm/train-balanced-sarcasm.csv")
data=pd.DataFrame(data)
X=data.iloc[:50000,1]
Y=data.iloc[:50000,0]
X=np.asarray(X)
Y=np.asarray(Y)
X=X.astype(str)
G=[one_hot(h,500) for h in X]
outputs=Y  
print(G)
print(outputs)
```

```python
padded=pad_sequences(G,maxlen=40,padding='post')
print(padded)
```

```python
model=Sequential()
model.add(Embedding(500,12,input_length=40,name='embed'))
model.add(SimpleRNN(30))
model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
padded=tensorflow.convert_to_tensor(padded)
testing="You do know west teams play against west teams more than east teams right?"
testing=one_hot(testing,500)
print(testing)
test=[]
for h in range(0,40):
  if h<len(testing):
   test.append(testing[h])
  else:
    test.append(0) 
test=[test]    
outputs=tensorflow.convert_to_tensor(outputs)
model.fit(padded,outputs,epochs=25)
q=model.predict(test,batch_size=3)
if q[0][0]>0.5:
  print("No sarcasm detected")
else:
  print("Sarcasm detected")   
```
